Creando series 

```python
#crear diferentes tipos de datos
labels = ['a','b','c'] #lista de etiquetas
my_list = [10,20,30] # lista con valores
arr = np.array([10,20,30]) #convertir lista de valores en arreglo numpy
d = {'a':10,'b':20,'c':30} #creación de un diccionario
```

Desde Listas

```python
#Convertir una lista en series utilizando el método pd.Series
#Observe que se crean los nombres con las posiciones de cada elemento
pd.Series(data=my_list)
```

```python
#También se puede escribir la lista directamente dentro de pd.Series()
pd.Series([10,20,20])
```

```python
#Convertir una lista en series utilizando pd.Series()
#Se puede ingresar el nombre de las posiciones
pd.Series(data=my_list, index=labels)
```

```python
#No es obligatorio escribir la palabra 'data'= dentro del argumento, tampoco el 'index'=
pd.Series(my_list, labels)
```

**Series**

A series is a sequence of data values. If a DataFrame is a table, a series is a list.

```python
pd.Series([1,2,3,4,])
```

A series is a single column of a DataFrame, you can assign row labels to the Series the same way as before, using and index parameter. Note that a series does not have a column name, it only have an overall name:

```python
pd.Series([30,35,40], index=['2015 Sales','2016 Sales','2017 Sales'], name= 'Product A')
```

You can think about a DataFrame as actually being a bunch of Series glued together.

Manipulating the index

```python
data.set_index("Local Customer Name")
```

```python
dataloc = data.loc[data["Customer Branch"] =="Colombiana De Comercio - Medellin"]
dataloc
```

```python
data.loc[(data["Customer Branch"]=="Colombiana De Comercio - Medellin" ) & (data["Shipment GIV Value"]>=60000) & (data["Shipment GIV Value"]<=100000)]
# filtering with loc, that means using index labels, it is very useful with multiple conditions
# df.loc[df["column label"]=="some string"] 
# df.loc[df["column labe"]]==some value] 
#With multiple conditions, we can include parenthesis at the start and at the end of each condition
```

```python
# Also working with multiple loc conditions using OR clause
data2 = data.loc[(data["Customer Branch"]=="Colombiana De Comercio - Medellin") | (data["Shipment GIV Value"]>=60000)]
data2
```

```python
new_df = data.loc[data['Customer Branch'].notnull()]
new_df.head()
```
